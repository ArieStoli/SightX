### data import and conversion

In this part I will convert the data from a JSON format into the following Pandas dataframes:
1. Movie_frames_df: stores the data for each frame, except for objects.
2. Movie_metadata_df: stores the first row of each movie - information about the movie itself.
3. Object_frames_df: stores the information about the objects in the movies, where and when they appear, and their different attributes.

```python
import json
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
json_dicts = []
for json_filename in os.listdir('./data'):
    with open(os.path.join('./data/',json_filename),'rb') as json_file:
        json_dicts.append(json.loads(json_file.read()))
```

### Building the dataframes:

```python
object_col_names = []

def build_object_dict(objs, object_list, movie, frame):
    temp = {}
    for obj in objs:
        temp['movie'] = movie
        temp['frame'] = frame
        if 'attributes' in obj.keys():
            for k,v in obj['attributes'].items():
                temp[k] = v
                if k not in object_col_names:
                    object_col_names.append(k)
            
            obj.pop('attributes')
            for k,v in obj.items():
                temp[k] = v
                if k not in object_col_names:
                    object_col_names.append(k)
            object_list.append(temp)
        
    return object_list
```

```python
movie_frames = []
object_frames = []
for i in range(len(json_dicts)):
    for j in range(len(json_dicts[i]))[1:]:
        json_dicts[i][j]['movieID'] = i
        if json_dicts[i][j]['attributes'] is not None:
            json_dicts[i][j]['ignore_frame'] = json_dicts[i][j]['attributes']['ignore_frame']
            json_dicts[i][j].pop('attributes')
        # print (json_dicts[i][j]['attributes'])
        object_frames = build_object_dict(json_dicts[i][j]['objects'],
                                              object_frames,
                                              i,
                                              j)
        if json_dicts[i][j]['objects'] is not None:
            json_dicts[i][j].pop('objects')
        movie_frames.append(json_dicts[i][j])
        
movie_frames_df = pd.DataFrame(movie_frames)
# movie_objects_df = pd.DataFrame(object_frames)
movie_frames_df
```

The table above shows the data values for each frame in all of the 5 movies. There isn't much to add, since there isn't a large amount of variation.

---

```python
objects_df = pd.DataFrame(object_frames)

objects_df
```

### Sub-task 3: taking a look into the movie frames data:
The main thing we want to see is the amount irrelevant frames (ignore_frame = True) there are out of all frames in a movie:

```python
ignore_counts = np.empty(shape=(len(json_dicts)), dtype=object)
for i in range(len(json_dicts)):
    ignore_counts[i] = {'movie': i,
                        'Num. of frames': movie_metadata_df.loc[i]['frames'],
                        'Num. of ignore': len(movie_frames_df[movie_frames_df['ignore_frame'] == True][movie_frames_df['movieID'] == i])
                        }
    ignore_counts[i]['precentage'] = '{:.3f}%'.format((ignore_counts[i]['Num. of ignore'] / ignore_counts[i]['Num. of frames']))

pd.DataFrame(ignore_counts.tolist())
```

It looks like there is only a small amount of rejects. If it was larger than 5% it would have been more difficult to dismissed them from dataset, since it'll leave large gaps in the video.

### Sub-task 4: taking a look into the Objects in the frames:
Here we will focus on the distribution of objects for each movie. This will be shown in the following bar chart.

```python
fig, ax = plt.subplots(5,1, figsize=(10,20), sharey='all',
                       gridspec_kw=dict(left=0.1, right=0.9,
                                        bottom=0.0005, top=0.99995))
for i in range(len(json_dicts)):
    objects, counts = np.unique(objects_df[objects_df['movie'] == i]['nm'], return_counts=True)
    ax[i].bar(objects, counts, color=['r','orange','k','b', 'g', 'm', 'y'])
    ax[i].set_title(f'object distribution for movie {i}')
    ax[i].set_xlabel('object nm')
    ax[i].set_ylabel('occured in frames')
plt.show()
```

We can see that objects' appearance is not distributed equaly. This is noticeable on all 5 ocassions, but especially in movie 3, where object 0.7 had appeared almost 6 times more than the second most common object.

We can also see that objects 0.1 and 0.2 are almost not featured in movies 3-4. Maybe the explanation is that the probability of the objects to appear together is small.
